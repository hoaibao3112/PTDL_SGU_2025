#### 2. Các thao tác trên dãy số a
(a) Bình phương các phần tử trong dãy số <br/>
(b) Tìm số lượng các phần tử của dãy số <br/>
(c) In ra giá trị lớn nhất, giá trị nhỏ nhất và giá trị trung bình <br/>
(d) Tính phương sai của dãy số $\sigma^2 = \frac{\sum{\left(a_i - \mu\right)^2}}{N - 1}$ với $\mu = \frac{\sum{a_i}}{N}$ và độ lệch chuẩn $var = \sqrt{\sigma^2}$

```python
import numpy as np
import math

# Dãy số đầu vào
a = [120, 183, 160, 175, 145, 162, 190, 160, 152, 162]

# a. Bình phương các phần tử trong dãy số
a_squared = [x**2 for x in a]
print("a. Binh phuong a: ", a_squared)

# b. Tìm số lượng các phần tử của dãy số
length = len(a)
print("b. Do dai a: ", length)

# c. In ra giá trị lớn nhất, giá trị nhỏ nhất và giá trị trung bình
max_val = max(a)
min_val = min(a)
mean_val = sum(a) / len(a)
print("c. Gia tri lon nhat: [{}], gia tri nho nhat: [{}], gia tri trung binh [{:.1f}]".format(max_val, min_val, mean_val))

# d. Tính phương sai và độ lệch chuẩn
# Tính giá trị trung bình μ
mu = sum(a) / len(a)

# Tính tổng bình phương độ lệch
sum_squared_deviation = sum((x - mu)**2 for x in a)

# Tính phương sai (σ²)
variance = sum_squared_deviation / (len(a) - 1)

# Tính độ lệch chuẩn (var = √σ²)
std_deviation = math.sqrt(variance)

print("d. Phuong sai: [{: .2f}] va do lech chuan: [{: .2f}]".format(variance, std_deviation))
```

 $$
 y = f\left(x| \mu, \sigma\right) = \frac{1}{\sigma\sqrt{2\pi}}exp\left(-\frac{\left(x - \mu\right)^2}{2\sigma^2}\right)
 $$
 nơi mà $x \in [\mu - 5\sigma, \mu + 5\sigma]$ <br/>
 (c) Sinh ngẫu nhiên dãy số có 100000 phần tử theo phân bố chuẩn $N(\mu = 1, \sigma^2 = 1.0)$ và vẽ đồ thị tần số xuất hiện <br/> <br/>

```python
# b. Sinh ngẫu nhiên N điểm (x, y) với phân phối chuẩn
mu, sigma = 1, 1.0
N = 1000  # Số điểm

# Tạo các điểm x trong khoảng [μ-5σ, μ+5σ]
x_range = np.linspace(mu - 5*sigma, mu + 5*sigma, N)

# Tính giá trị y theo hàm mật độ xác suất chuẩn
y_pdf = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x_range - mu)/sigma)**2)

# Vẽ đồ thị
plt.figure(figsize=(8,6))
plt.plot(x_range, y_pdf, 'r-', lw=2)
plt.xlabel('x')
plt.ylabel('f(x|μ,σ)')
plt.title('Hàm mật độ xác suất phân phối chuẩn N(μ=1, σ²=1.0)')
plt.grid(True)
plt.savefig("4b.png")
plt.show()

# In thông tin
print("b. Sinh ngẫu nhiên N điểm (x, y) với phân phối chuẩn")
print(f"+ Số điểm N: {N}")
print(f"+ Khoảng x: từ {mu-5*sigma:.2f} đến {mu+5*sigma:.2f}")
print(f"+ Giá trị y lớn nhất: {np.max(y_pdf):.4f} tại x = {mu}")
```

```python
# c. Sinh ngẫu nhiên dãy số có 100000 phần tử theo phân bố chuẩn
normal_data = np.random.normal(mu, sigma, 100000)

# Vẽ histogram
plt.figure(figsize=(8,6))
plt.hist(normal_data, bins=50, density=True, alpha=0.7, color='blue')
plt.xlabel('Giá trị')
plt.ylabel('Mật độ xác suất')
plt.title('Phân bố chuẩn - 100000 phần tử')
plt.grid(True)
plt.savefig("4c.png")
plt.show()

# Tính và hiển thị thống kê cơ bản
print("c. Sinh ngẫu nhiên theo phân bố chuẩn")
print(f"+ Trung bình: {np.mean(normal_data):.4f}")
print(f"+ Độ lệch chuẩn: {np.std(normal_data):.4f}")
print(f"+ Min: {np.min(normal_data):.4f}")
print(f"+ Max: {np.max(normal_data):.4f}")
```

#### 4. Giải hệ phương trình tuyến tính
Giải hệ phương trình tuyến tính sau:
$$
\left\{\begin{matrix}
        4x_1  + 3x_2 - 5x_3 = 2 \\
        -2x_1 - 4x_2 + 5x_3 = 5 \\
        8x_1 + 8x_2 = -3 \\
\end{matrix}\right.
$$

```python
import numpy as np

# Ma trận A
A = np.array([[4, 3, -5],
              [-2, -4, 5],
              [8, 8, 0]])

# Vector y
y = np.array([2, 5, -3])

# Giải hệ phương trình Ax = y
x = np.linalg.solve(A, y)

# Kiểm tra độ chính xác
y_calculated = A @ x
diff = np.linalg.norm(y - y_calculated)

# In kết quả
print(f"Ma tran A: \n{A}")
print(f"Vector y: {y}")
print(f"Nghiem x: {x}")
print(f"Kiem tra: y = Ax = {y_calculated} => ||y - y||_2 = {diff:.2f}")
```

### Bài 3. Tính gần đúng số pi dùng phương pháp Monte Carlo

Vẽ một đường tròn $C$ có bán kính $r$ nội tiếp một hình vuông $C$, và đặt ngẫu nhiên $N$ chấm lên hình vuông. Tỉ lệ các chấm nằm trong hình tròn trên tổng số chấm xấp xỉ bằng diện tích của hình tròn chia cho hình vuông như sau:
$$
\frac{N_C}{N} \approx  \frac{\mathbb{S}_C}{\mathbb{S}_N}
$$
nơi mà $N_C$ là các chấm trong hình tròn, $N$ là tổng số chấm, $\mathbb{S}_C$ và $\mathbb{S}_N$ lần lượt là diện tích của hình tròn và hình vuông.
$$ \mathbb{S}_C = \pi * r^2 $$
$$ \mathbb{S}_N = (2*r)^2 = 4r^2 $$
Do đó, công thức trên trở thành:
$$
\frac{N_C}{N} \approx \frac{\pi * r^2}{4r^2} = \frac{\pi}{4}
$$
Suy ra, ta có giá trị số $\pi$ được tính như sau:
$$
\pi \approx \frac{4N_C}{N}
$$

Các bạn hãy lập trình tính sấp xỉ số $\pi$ dùng phương pháp Monte Carlo. Cho biết các sai số với $N = 100$, $N = 10000$, và $N = 1000000$. Vẽ đồ thị minh họa cho chương trình.

**Hướng dẫn**: Viết chương trình phát sinh $N$ điểm $(x_i, y_i)$ với $x_i, y_i \in [-1, -1]$. $(x_i, y_i)$ thuộc hình tròn tâm $(0, 0)$ bán kính 1 nội tiếp hình vuông có tâm tại $(0, 0)$ độ dài cạnh 2 khi và chỉ khi $x_i^2 + y_i^2 <= 1$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Định nghĩa các giá trị N để thử nghiệm
N_values = [100, 1000, 100000]

# Vòng lặp để tính toán và hiển thị kết quả cho từng N
for N in N_values:
    # Tạo các điểm ngẫu nhiên trong hình vuông [-1, 1] x [-1, 1]
    x = np.random.uniform(-1, 1, N)
    y = np.random.uniform(-1, 1, N)

    # Đếm số điểm nằm trong đường tròn đơn vị (x^2 + y^2 <= 1)
    inside_circle = x**2 + y**2 <= 1
    Nc = np.sum(inside_circle)

    # Ước lượng giá trị π
    pi_approx = 4 * Nc / N
    print(f"N = {N}, Giá trị π ước lượng: {pi_approx}")

    # Vẽ biểu đồ minh họa cho N = 1000
    if N == 1000:
        fig, ax = plt.subplots(figsize=(6, 6))
        circle = plt.Circle((0, 0), 1, color='g', fill=False)
        ax.add_patch(circle)
        ax.plot(x[inside_circle], y[inside_circle], 'b.', label='Trong vòng tròn')
        ax.plot(x[~inside_circle], y[~inside_circle], 'r.', label='Ngoài vòng tròn')
        ax.set_xlim(-1, 1)
        ax.set_ylim(-1, 1)
        ax.set_aspect('equal')
        ax.legend()
        plt.title(f'Monte Carlo với N = {N}')
        plt.savefig(f'pi_N{N}.png')
        plt.show()

# Vẽ minh họa đường mũi tên
x_arrow = 0.75
y_arrow = np.sqrt(1 - x_arrow**2)
fig, ax = plt.subplots(figsize=(6, 6))
circle = plt.Circle((0, 0), 1, color='g', fill=False)
ax.add_patch(circle)
ax.arrow(0, 0, x_arrow, y_arrow, head_width=0.1, head_length=0.1, fc='r', ec='r')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')
plt.title('Minh họa điểm trong vòng tròn')
plt.savefig('pi_arrow.png')
plt.show()
```

```python
import numpy as np

def calc_pi_monte_carlo(n=100):
    pi = 0

    # Sinh ngẫu nhiên n điểm trong hình vuông [0, 1] x [0, 1]
    x = np.random.rand(n)
    y = np.random.rand(n)

    # Đếm số điểm nằm trong phần tư đường tròn đơn vị (x^2 + y^2 <= 1)
    inside_circle = x**2 + y**2 <= 1
    num_inside = np.sum(inside_circle)

    # Ước lượng π: vì đây là phần tư vòng tròn, nhân 4 để được π đầy đủ
    pi = 4 * num_inside / n

    return pi

# Kiểm tra hàm
print(calc_pi_monte_carlo())  # Kết quả với n=100
print(calc_pi_monte_carlo(1000))  # Kết quả với n=1000
```

```python
import math

print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
```
